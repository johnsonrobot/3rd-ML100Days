# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

def relu(X):
    return np.maximum(0,X)
x = plt.linspace(-10, 10, 100)
plt.plot(x, relu(x), 'b', label='linspace(-10, 10, 10)')
plt.grid()
plt.title('ReLu function')

plt.text(3.6, 0.8, r'$f({x})=\max(0, {x})$', fontsize=15)

plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
plt.gca().yaxis.set_major_locator(plt.MultipleLocator(1))

plt.show()
```

```python
def drelu(x):
    x[x <= 0] = 0
    x[x > 0] = 1
    return x
x1 = plt.linspace(-10, 10, 100)
x2 = plt.linspace(-10, 10, 100)
plt.plot(x1, drelu(x2), 'b', label='linspace(-10, 10, 100)')
plt.grid()
plt.title('ReLu function')

plt.text(3, 0.8, 'if x > 0, x = 1\nif x <= 0, x = 0', fontsize=15)

plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
plt.gca().yaxis.set_major_locator(plt.MultipleLocator(1))

plt.show()
```
